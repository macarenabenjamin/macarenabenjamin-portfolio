**2) Se extraen los datos de DATE_TRX, OPEN_PRICE, HIGH_PRICE, LOW_PRICE, CLOSE_PRICE, ADJ_CLOSE_PRICE, VOLUME de los activos arriba mencionados a traves de la API de Yahoo Finance**

```
#Primero nos conectamos con la API de Yahoo Finance

#se importan los datos a través de la API
import yfinance as yf
from pandas_datareader import data as pdr
#conexión a la API
yf.pdr_override()
```

```
# Lista de activos
activos = ["MSFT", "INTC", "NKE.VI", "CRM", "HD", "MMM", "KO", "AAPL", "HON", "PG",
           "MCD", "WMT", "IBM", "BA", "CSCO", "VZ", "AXP", "DOW", "GS", "JPM",
           "WBA", "VISA.VI", "AMGN", "MRK", "CAT", "PA9.F", "JNJ", "DIS", "UNH", "CVX"]

# Fecha de inicio y fin
start_date = "2024-03-18"
end_date = "2024-03-23" #para incluir al 22/03/24

# Obtener datos para cada ticker
data = {}
for activo in activos:
    data[activo] = yf.download(activo, start=start_date, end=end_date)

# Crear DataFrame con los datos necesarios
datos_YF = pd.DataFrame(columns=['DATE_TRX', 'OPEN_PRICE', 'HIGH_PRICE', 'LOW_PRICE', 'CLOSE_PRICE', 'ADJ_CLOSE_PRICE', 'VOLUME'])

for activo, df in data.items():
    if not df.empty:
        df.reset_index(inplace=True)
        df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
        df.columns = ['DATE_TRX', 'OPEN_PRICE', 'HIGH_PRICE', 'LOW_PRICE', 'CLOSE_PRICE', 'ADJ_CLOSE_PRICE', 'VOLUME']
        df['SYMBOL_YAHOO_FINANCE'] = activo
        datos_YF = pd.concat([datos_YF, df], ignore_index=True)
```

```
# Mostrar resultados
datos_YF.head(10)
```

**3) Creación de base de datos SQL**

```
import sqlite3 as sql

#se crea a una base sql dandole un nombre
conexión = sql.connect("DJII.db")

```

```
import pandas.io.sql as pd_sql

#se almacena el DataFrame como tabla en la base creada
pd_sql.to_sql(datos_YF,'DATOS_YF',conexión, index=False)
#donde datos_YF es el el DataFrame, 'DATOS_YF' es el nombre de la tabla creada en la base,
# conexión es la conexión a la base, index=False para no almacenar el índice de Python
```
