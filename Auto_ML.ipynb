## **Introduccion Dataset**

Open Electricity es una plataforma abierta para rastrear o dar seguimiento del consumo de electricidad en Australia segun el tipo de energia.
Los datos de consumo se presentan en la unidad GWh que significa gigavatio-hora y es una unidad de energía que mide la cantidad de electricidad producida, consumida o almacenada durante una hora.

El data set esta compuesto por:
- **date**: Fecha

Tipos de energia medidos en consumo GWh:

- **Coal (Black) - GWh**: Carbón (negro) - GWh
- **Bioenergy (Biogas) - GWh**: Bioenergía (biogás) - GWh
- **Bioenergy (Biomass) - GWh**: Bioenergía (biomasa) - GWh
- **Distillate - GWh**: Destilado - GWh
- **Gas (Steam) - GWh**: Gas (vapor) - GWh
- **Gas (CCGT) - GWh**: Gas (CCGT - ciclo combinado de gas) - GWh
- **Gas (OCGT) - GWh**: Gas (OCGT - turbina de gas de ciclo abierto) - GWh
- **Gas (Reciprocating) - GWh**: Gas (reciprocante) - GWh
- **Gas (Waste Coal Mine) - GWh**: Gas (de mina de carbón residual) - GWh
- **Battery (Discharging) - GWh**: Batería (descargando) - GWh
- **Hydro - GWh**: Hidroeléctrica - GWh

- **Solar (Utility) - GWh**: Solar (a gran escala) - GWh

Volumen de emision de carbono por tipo de energia medido en toneladas de dióxido de carbono equivalente:

# **Extraccion de datos**

Los datos se extraen de la pagina de Open Electricity https://explore.openelectricity.org.au/energy/nem/?range=all-12-mth-rolling&interval=1M&view=discrete-time&group=Detailed de manera mensual desde Diciembre 1999 hasta Septiembre 2024 como csv y se suben al drive para poder llamarlos.


```python
#Se abre google drive
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

#Se llama al archivo que contiene informacion mensual desde Diciembre 1999 hasta Septiembre 2024
file_path = '/content/drive/MyDrive/19991201 Open Electricity.csv'

#Leer el archivo CSV en un DataFrame
df = pd.read_csv(file_path)

#Mostrar DataFrame
df.head(300)
```

```python
#Se ven las columnas
df.columns
```

# **Transformacion y limpieza de datos**

```python
#Como el objetivo del trabajo es predecir la demanda de energia se suma en consumo de los distintos tipos de energia en una nueva variable que se llame 'TotalEnergy'

import pandas as pd

#Selección de columnas de consumo de energía en GWh
energy_columns = [
    'Battery (Charging) -  GWh', 'Pumps -  GWh', 'Coal (Brown) -  GWh',
    'Coal (Black) -  GWh', 'Bioenergy (Biogas) -  GWh', 'Bioenergy (Biomass) -  GWh',
    'Distillate -  GWh', 'Gas (Steam) -  GWh', 'Gas (CCGT) -  GWh',
    'Gas (OCGT) -  GWh', 'Gas (Reciprocating) -  GWh', 'Gas (Waste Coal Mine) -  GWh',
    'Battery (Discharging) -  GWh', 'Hydro -  GWh', 'Wind -  GWh',
    'Solar (Utility) -  GWh', 'Solar (Rooftop) -  GWh'
]

#Crear la nueva columna sumando las seleccionadas
df['TotalEnergy'] = df[energy_columns].sum(axis=1)

df.head()
```

```python
#Se verifica el shape del set de datos
df.shape
```

```python
#Se verifica el tipo de datos
df.dtypes
```

```python
#Se convierte la columna 'date' en tipo datetime
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df.head()
```

```python
#Se verifica
df.dtypes
```

```python
#Se verifican nulos
df.isnull().sum()
```

No hay valores faltantes

# **Preparacion del set de datos**

```python
# Se normalizan los datos quitando la columna 'date'

# Lista de columnas que se desea normalizar (todas menos 'date')
numeric_columns = df.columns.drop('date')

# Se normaliza cada columna en el DataFrame y transforma a float32
for c in numeric_columns:
    df[c] = ((df[c] - df[c].min()) / (df[c].max() - df[c].min())).astype('float32')

# Mostrar las primeras filas del DataFrame normalizado
df.head()
```

```python
#Se verifica el tipo de datos
df.dtypes
```

```python
#Se grafican los consumos segun tipo de energia
import matplotlib.pyplot as plt
plt.figure(figsize=(20, len(energy_columns) * 2 ))
for i, c in enumerate(energy_columns):
    plt.subplot(len(energy_columns), 1, i + 1)
    plt.plot(df['date'], df[c])
    plt.title(c, y=1, loc='left')
    plt.xlabel('Fecha')
    plt.ylabel('Consumo de Energía (GWh)')
plt.tight_layout()
plt.show()
```

```python
#Se separan en subconjutos de train y test
train = df[df['date'] < '2019-01-01']
test  = df[df['date'] >= '2019-01-01']
train.shape, test.shape
```

```python
#Se quita la variable 'date'porque no es número
cols = df.columns.drop(['date'])

window_size = 3 #Hiperparametro

#Se define el subconjunto X en train y test
X_train = pd.concat([pd.concat([train[c].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"{c}_{w}" for w in range(1, window_size+1)]) for c in cols], axis=1)
X_test  = pd.concat([pd.concat([test[c].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"{c}_{w}" for w in range(1, window_size+1)]) for c in cols], axis=1)
X_test.head()

#Se define el subconjunto y en train y test

y_train = train['TotalEnergy']
y_test  = test['TotalEnergy']
X_train.head()
```

```python
#Se verifica formato (n_samples, n_timesteps, n_features)
X_train = X_train.values.reshape( (X_train.shape[0], X_train.shape[1], 1) )
X_test  = X_test.values.reshape( (X_test.shape[0], X_train.shape[1], 1) )
X_train.shape, X_test.shape
```

# **Modelo Keras sin AutoML**

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    #capa input
    tf.keras.layers.InputLayer(input_shape=[X_train.shape[1], X_train.shape[2]]),

    #capas de LSTM
    tf.keras.layers.LSTM(12, return_sequences=True), #12 porque el ciclo es de 12 meses - hiperparámetro
    tf.keras.layers.LSTM(window_size),

    #capa densa
    tf.keras.layers.Dense(1, activation='linear'),  #Salida continua porque es un problema de regresion
])

model.compile(loss="mean_squared_error", optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)) #se agrega un learning rate


history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=50, verbose=True)
```

```python
# predicciones
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
# RMSE
import numpy as np
np.sqrt(np.mean((test_predict.flatten() - y_test)**2))
```

```python
#Se grafica la prediccion
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['TotalEnergy'], label="TotalEnergy")
plt.plot(test['date'][window_size:], test_predict[window_size:], label="Prediccion")
plt.xlabel("Fecha")
plt.ylabel("Total Consumo de Energia")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
import keras_tuner as kt

#Se define el modelo

def build_model(hp):
    model = tf.keras.models.Sequential()

    #Capa de entrada
    model.add(tf.keras.layers.InputLayer(input_shape=[X_train.shape[1], X_train.shape[2]]))

    #Capas LSTM con hiperparametros
    model.add(tf.keras.layers.LSTM(
        units=hp.Int('lstm_units_1', min_value=8, max_value=16, step=2),
        return_sequences=True
    ))
    model.add(tf.keras.layers.LSTM(
        units=hp.Int('lstm_units_2', min_value=2, max_value=6, step=1)
    ))

    #Capa densa
    model.add(tf.keras.layers.Dense(1, activation='linear'))

    #Compilacion del modelo con hiperparametros de tasa de aprendizaje
    model.compile(
        loss="mean_squared_error",
        optimizer=tf.keras.optimizers.Adam(
            learning_rate=hp.Choice('learning_rate', values=[0.001, 0.005, 0.01])
        )
    )

    return model
```

```python
#Configuracion de Keras Tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=20,
    factor=3,
    directory='my_dir',
    project_name='keras_tuner_lstm'
)

#Busqueda de hiperparametros
tuner.search(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=50)

#Recuperar el mejor modelo
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()
```

```python
#Imprimir los mejores hiperparametros
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print("Mejores hiperparámetros encontrados:")
print(f"Unidades LSTM 1: {best_hps.get('lstm_units_1')}")
print(f"Unidades LSTM 2: {best_hps.get('lstm_units_2')}")
print(f"Tasa de aprendizaje: {best_hps.get('learning_rate')}")
```

```python
best_trial = tuner.oracle.get_best_trials(num_trials=1)[0]
print(f"El mejor val_loss durante la búsqueda fue: {best_trial.metrics.get_last_value('val_loss')}")
```

```python
#Entrenamiento final del mejor modelo
history = best_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=50, verbose=True)
```

```python
#predicciones
train_predict_best = best_model.predict(X_train)
test_predict_best = best_model.predict(X_test)
```

```python
#RMSE
import numpy as np
np.sqrt(np.mean((test_predict_best.flatten() - y_test)**2))
```

```python
import matplotlib.pyplot as plt
#Se grafica la prediccion
plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['TotalEnergy'], label="TotalEnergy")
plt.plot(test['date'][window_size:], test_predict_best[window_size:], label="Prediccion")
plt.xlabel("Fecha")
plt.ylabel("Total Consumo de Energia")
plt.legend()
plt.show()
```
