# **Generacion de imagenes y Clasificacion**

```python
import os
import pandas as pd
import mplfinance as mpf
from PIL import Image

#Funcion para generar los candlestick chart, recortar y ajustar la imagen
def generar_imagen_candlestick(datos, archivo):
    mpf.plot(
        datos,
        type='candle',
        style=mpf.make_mpf_style(gridstyle='', facecolor='white'),
        title='',
        ylabel='',
        volume=True,
        ylabel_lower='',
        show_nontrading=True,
        savefig=archivo
    )

    #Abrir, recortar y ajustar la imagen
    im = Image.open(archivo)
    im2 = im.crop((145, 70, 720, 473))
    im3 = im2.resize((400, 400))
    im3.save(archivo)

#Funcion para clasificar y guardar las imágenes en carpetas
def clasificar_y_guardar(precio_cierre_anterior, precio_cierre_actual, archivo):

     #Calcular el porcentaje de variacion de precio basado en el precio actual
    cambio_porcentual = (precio_cierre_actual - precio_cierre_anterior) / precio_cierre_actual * 100


    #Clasificacion
    if cambio_porcentual > 1: #Sugerencia de Compra
        carpeta = "Comprar"
    elif cambio_porcentual < -1: #Sugerencia de Venta
        carpeta = "Vender"
    else:  #Sugerencia de Mantener
        carpeta = "Hold"

    #Crear la carpeta si no existe
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)

    #Mover la imagen a la carpeta correspondiente
    nuevo_archivo = os.path.join(carpeta, os.path.basename(archivo))
    os.rename(archivo, nuevo_archivo)
    print(f"Image moved to: {nuevo_archivo}")

#Loop para generar y clasificar las imagenes para cada dia habil
for fecha_actual in datos.index:
    #Extraer los ultimos 5 datos disponibles antes de la fecha actual (no incluir la fecha actual)
    ultimos_datos = datos.loc[:fecha_actual].iloc[:-1].tail(5)

    #Verificar que hay al menos 5 dias de datos disponibles
    if len(ultimos_datos) < 5:
        print(f"No hay suficientes datos para generar la imagen para {fecha_actual}.")
        continue  #Si no hay suficientes datos, saltar a la siguiente fecha


    #Extraer el precio de cierre del dia actual
    precio_cierre_actual = datos.loc[fecha_actual, 'Close']  #Precio de cierre del dia actual

    #Extraer el precio de cierre del ultimo dia de los 5 dias anteriores
    precio_cierre_anterior = ultimos_datos['Close'].iloc[-1]  #Ultimo dia de los 5 seleccionados

    #Definir el nombre de la imagen
    nombre_archivo = f'BMBL_{fecha_actual.strftime("%Y-%m-%d")}.png'

    #Generar la imagen del candlestick
    generar_imagen_candlestick(ultimos_datos, nombre_archivo)

    #Clasificar y guardar en la carpeta
    clasificar_y_guardar(precio_cierre_anterior, precio_cierre_actual, nombre_archivo)
```

#**Generador de entrenamiento y validacion**

```python
import os
import tensorflow as tf
from google.colab import drive

#Abrir Google Drive
drive.mount('/content/drive')

#Tamaño de lote
batch_size = 128

#Ruta a la carpeta de imagenes en Google Drive
carpeta_destino_base = '/content/drive/My Drive/BMBL_Clasificadas/'

#Crear un ImageDataGenerator para preprocesar las imagenes
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    zoom_range=1.0,
    validation_split=0.2  #Se agrega el porcentaje de validacion
)

#Crear un generador de entrenamiento desde las carpetas
train_generator = train_datagen.flow_from_directory(
    carpeta_destino_base,  #Directorio donde estan las imagenes
    target_size=(100, 100),  #Redimensionar las imagenes a 100x100
    #color_mode='grayscale',
    batch_size=batch_size,
    classes=['Comprar', 'Vender', 'Hold'],  #Clases
    subset='training'  #Entrenamiento
)

#Crear un generador de validacion
validation_generator = train_datagen.flow_from_directory(
    carpeta_destino_base,
    target_size=(100, 100),
    #color_mode='grayscale',
    batch_size=batch_size,
    classes=['Comprar', 'Vender', 'Hold'],
    subset='validation'  #Validacion
)
```

# **Modelo**

```python
model = tf.keras.models.Sequential([

    # Entrada
    tf.keras.layers.Input(shape=(100, 100, 3)), # 3 en color, 1 es grayscale

    # Convolución 1
    tf.keras.layers.Conv2D(16, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolución 2
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolución 3
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolución 4
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolución 5
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Achatamiento
    tf.keras.layers.Flatten(),

    # Capas densas
    tf.keras.layers.Dense(100, activation='relu'),

    # cantidad de neuronas de salida, una por clase (como son 3 clases, se coloca 3)
    tf.keras.layers.Dense(3, activation='softmax')

])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#model.summary()

n_epochs = 10
history = model.fit(train_generator, epochs=n_epochs, verbose=2, validation_data=validation_generator )
```

# **Optimizacion**

```python
#Se aumento la cantidad de epocas para ver si mejora el modelo
n_epochs = 30
history = model.fit(train_generator, epochs=n_epochs, verbose=2, validation_data=validation_generator )
```
